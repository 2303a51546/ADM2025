```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter


np.random.seed(42)


N = 1000

max_calls = 7


calls_per_day = np.random.randint(0, max_calls + 1, size=N)


counter = Counter(calls_per_day)
pmf = [counter[i] / N for i in range(max_calls + 1)]
x_values = np.arange(max_calls + 1)


cdf = np.cumsum(pmf)


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x_values, pmf, color='blue', alpha=0.7, width=0.6)
plt.title('Probability Mass Function (PMF)')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.xticks(x_values)
plt.grid(axis='y')


plt.subplot(1, 2, 2)
plt.step(x_values, cdf, where='mid', color='orange', linewidth=2)
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Number of Calls')
plt.ylabel('Cumulative Probability')
plt.xticks(x_values)
plt.grid()


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats


N = 100
p = 0.25


k_exactly_20 = 20
prob_exactly_20 = stats.binom.pmf(k_exactly_20, N, p)


k_at_least_30 = 30
prob_at_least_30 = 1 - stats.binom.cdf(k_at_least_30 - 1, N, p)


expected_patients = N * p


print(f"Probability that exactly {k_exactly_20} patients need surgery: {prob_exactly_20:.4f}")
print(f"Probability that at least {k_at_least_30} patients need surgery: {prob_at_least_30:.4f}")
print(f"Expected number of patients requiring surgery: {expected_patients:.2f}")
```
