```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('employee_data_250-2.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=20, kde=False)
plt.title('Distribution of Salary')
plt.xlabel('Salary (USD)')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Experience (Years)"], y=df["Salary"], color="blue", alpha=0.6)
sns.regplot(x=df["Experience (Years)"], y=df["Salary"], scatter=False, color="red")

plt.xlabel("Experience (Years)")
plt.ylabel("Salary (USD)")
plt.title("Experience vs. Salary")
plt.show()

# Calculate correlation between Experience and Salary
correlation = df["Experience (Years)"].corr(df["Salary"])
print(f"Correlation between Experience and Salary: {correlation}")
```

```python
plt.figure(figsize=(10, 6))
df['Department'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
job_level_department = df.groupby(['Department', 'Job Satisfaction']).size().unstack()
job_level_department.plot(kind='bar', stacked=False)
plt.title('Number of Employees by Job Level Across Departments')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.legend(title='Job Satisfaction')
plt.show()
```
